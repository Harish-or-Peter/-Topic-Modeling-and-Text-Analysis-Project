# **Project Summary -**

**Topic Modeling on News Articles**

**Summary:**

Text data is a valuable source of information across various domains, but making sense of large volumes of unstructured text can be a daunting task. Topic modeling, a subfield of natural language processing (NLP), offers a powerful solution to this challenge. This summary explores a project focused on topic modeling and its applications, highlighting key methodologies, techniques, and results.

**Project Overview:**

The project's primary objective is to employ topic modeling techniques to extract meaningful topics or themes from a corpus of news articles. These topics can then be used for a range of applications, including document categorization, content recommendation, and understanding latent patterns within the data.

**Data Preparation:**

The first step involves data preprocessing. In this project, a collection of paragraphs was used as the raw text data. The text was cleaned to remove noise, which includes punctuation, stopwords, and special characters. Tokenization and lemmatization were performed to break down the articles into individual words and reduce words to their base form, respectively. This step is crucial for improving the quality of topic modeling.

**Feature Engineering:**

After data preprocessing, the next step is feature engineering. In this project, a TF-IDF (Term Frequency-Inverse Document Frequency) vectorizer was used to convert the text data into a numerical format suitable for modeling. TF-IDF assigns weights to words based on their importance within a document and across the entire corpus.

**Topic Modeling Algorithms:**

Three different topic modeling algorithms were explored in this project: Latent Dirichlet Allocation (LDA), Non-Negative Matrix Factorization (NMF), and Latent Semantic Analysis (LSA). These algorithms are widely used for uncovering hidden topics within text data.

1. **Latent Dirichlet Allocation (LDA):** LDA is a probabilistic generative model that assumes each document is a mixture of topics, and each word in a document is attributable to one of its document's topics. It estimates the distribution of topics in the corpus and the distribution of words in each topic.

2. **Non-Negative Matrix Factorization (NMF):** NMF factorizes the TF-IDF matrix into two lower-dimensional matrices, one representing topics and the other representing the distribution of topics in documents. It enforces non-negativity, which makes the resulting matrices interpretable.

3. **Latent Semantic Analysis (LSA):** LSA is based on singular value decomposition (SVD) and reduces the dimensionality of the TF-IDF matrix. It captures underlying patterns by preserving the most significant singular values.

**Hyperparameter Tuning:**

Hyperparameter tuning is a critical aspect of topic modeling. It involves finding the optimal number of topics (k) for each algorithm. Techniques like grid search and coherence scores were used to identify the best k value for each model. Coherence scores measure the interpretability of topics, helping to select models with more meaningful topics.


**Conclusion:**

In summary, this project demonstrated the power of topic modeling techniques in uncovering hidden themes within text data. By preprocessing the data, selecting appropriate algorithms, tuning hyperparameters, and evaluating model performance, meaningful topics were extracted and applied to various real-world tasks. The project showcases the versatility of topic modeling and its potential to extract actionable insights from unstructured text data.

## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Okay

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Null Hypothesis (H0):** The average article length in the "politics" category is equal to the average article length in the "entertainment" category.

**Alternative Hypothesis (H1):** The average article length in the "politics" category is not equal to the average article length in the "entertainment" category.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value

politics_lengths = df_cleaned[df_cleaned['category'] == 'politics']['word_count']
entertainment_lengths = df_cleaned[df_cleaned['category'] == 'entertainment']['word_count']

# Perform two-sample t-test
t_stat, p_value = stats.ttest_ind(politics_lengths, entertainment_lengths, equal_var=False)

# Display the t_stat
print("T-statistic:", t_stat)

# Display the p-value
print("P-Value:", p_value)

alpha = 0.05

if p_value < alpha:
    print('We reject the null hypothesis and conclude that the average article length in the "politics" category is not equal to the average article length in the "entertainment" category.')
else:
    print('We fail to reject the null hypothesis. There is no significant difference in average article lengths between the categories.')
````

##### Which statistical test have you done to obtain P-Value?

The statistical test employed to obtain the p-value is the **two-sample independent t-test assuming unequal variances**. This test assesses whether there is a significant difference in the means of two independent samples. In this case, the samples are the article lengths from the "politics" and "entertainment" categories. The t-test calculates a t-statistic that quantifies the difference between the sample means, considering the variability within each sample. The p-value, resulting from the t-test, indicates the probability of observing such a difference if the null hypothesis (no difference between means) is true. Comparing the p-value to a predetermined significance level helps determine whether to reject the null hypothesis or not. If the p-value is below the significance level, the null hypothesis is rejected, suggesting a significant difference in average article lengths between the categories.

##### Why did you choose the specific statistical test?

The choice of the two-sample independent t-test was driven by the nature of the comparison between article lengths in the "politics" and "entertainment" categories. This test is suitable for assessing differences in means between two independent groups, making it appropriate for comparing the average word counts of articles from distinct categories. The assumption of unequal variances accounts for potential variations in data distribution, ensuring accurate results even when sample sizes and variances differ. Additionally, the t-test aligns with the hypothesis testing framework, enabling a clear evaluation of whether observed differences are statistically significant or could have occurred due to random chance.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Null Hypothesis (H0):** The average article length in the "tech" category is equal to the average article length in the "business" category.

````python
# Perform Statistical Test to obtain P-Value

# Extract article lengths for 'tech' and 'business' categories
tech_lengths = df_cleaned[df_cleaned['category'] == 'tech']['word_count']
business_lengths = df_cleaned[df_cleaned['category'] == 'business']['word_count']

# Perform two-sample t-test
t_stat, p_value = stats.ttest_ind(tech_lengths, business_lengths, equal_var=False)

# Display the t_stat
print("T-statistic:", t_stat)

# Display the p-value
print("P-Value:", p_value)

alpha = 0.05

if p_value < alpha:
    print('We reject the null hypothesis and conclude that the average article length in the "tech" category is not equal to the average article length in the "business" category.')
else:
    print('We fail to reject the null hypothesis. There is no significant difference in average article lengths between the categories.')
````

##### Which statistical test have you done to obtain P-Value?

This test is chosen for comparing the means of two independent samples, in this case, the average article lengths from the "tech" and "business" categories. The t-test accounts for variations in sample sizes and variances between the groups. **Null Hypothesis (H0):** The average article length in the "sports" category is equal to the average article length in the "politics" category.

**Alternative Hypothesis (H1):** The average article length in the "sports" category is not equal to the average article length in the "politics" category.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value

# Extract article lengths for 'sports' and 'politics' categories
sports_lengths = df_cleaned[df_cleaned['category'] == 'sport']['word_count']
politics_lengths = df_cleaned[df_cleaned['category'] == 'politics']['word_count']

# Perform two-sample t-test
t_stat, p_value = stats.ttest_ind(sports_lengths, politics_lengths, equal_var=False)

# Display the t_stat
print("T-statistic:", t_stat)

# Display the p-value
print("P-Value:", p_value)

alpha = 0.05

if p_value < alpha:
    print('We reject the null hypothesis and conclude that the average article length in the "sports" category is not equal to the average article length in the "politics" category.')
else:
    print('We fail to reject the null hypothesis. There is no significant difference in average article lengths between the categories.')
````

#### 1. Expand Contraction

````python
# Expand Contraction

# Function to expand contractions using the contractions_dict
def expand_contractions(text):
    # Regular expression pattern to match contractions
    contractions_pattern = re.compile('({})'.format('|'.join(contractions_dict.keys())),
                                      flags=re.IGNORECASE | re.DOTALL)

    def expand_match(contraction):
        match = contraction.group(0)
        expanded = contractions_dict.get(match.lower())
        return expanded

    expanded_text = contractions_pattern.sub(expand_match, text)
    return expanded_text

# Apply the expand_contractions function to the "content" column
df_cleaned['content'] = df_cleaned['content'].apply(expand_contractions)
````

````python
# Remove Punctuations

# Function to remove punctuations from text
def remove_punctuations(text):
    # Create a translation table to remove punctuations
    translator = str.maketrans('', '', string.punctuation +'\n')

    # Apply the translation table to remove punctuations
    text_without_punctuations = text.translate(translator)
    return text_without_punctuations

# Apply the remove_punctuations function to the "content" column
df_cleaned['content'] = df_cleaned['content'].apply(remove_punctuations)

print('\n')
print('\n')
# Display the content of the first row
df_cleaned['content'][0]
````

````python
# Remove URLs & Remove words and digits contain digits

# Function to remove URLs from text
def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return re.sub(url_pattern, '', text)

# Function to remove words containing digits from text
def remove_words_with_digits(text):
    return ' '.join(word for word in text.split() if not any(char.isdigit() for char in word))

# Apply the remove_urls function to the "content" column
df_cleaned['content'] = df_cleaned['content'].apply(remove_urls)

# Apply the remove_words_with_digits function to the "content" column
df_cleaned['content'] = df_cleaned['content'].apply(remove_words_with_digits)

print('\n')
print('\n')
# Display the content of the first row
df_cleaned['content'][0]
````
